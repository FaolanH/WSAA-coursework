## Defining the assignment:

Using Deck of Cards API (https://deckofcardsapi.com/) an API that simulates dealing a deck of cards, write a program that "deals" (prints out) 5 cards.

Step 2: Get the deck_id

Step 3: With the deck_id you can get the cards (https://deckofcardsapi.com/api/deck/<<deck_id>>/draw/?count=2)

Step 4: Print the value and the suit of each card

Step 5: Save the file as assignment2-carddraw.ipynb

Additional (optional) step: Check if the user has drawn a pair, triple, straight, or all of the same suit and congratulate the user.
