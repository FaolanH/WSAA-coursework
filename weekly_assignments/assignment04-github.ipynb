##### I will be using the book.json file which uses a book by Andrew Beatty. 

```python
import json
```

```python
# Source: https://stackoverflow.com/questions/53077503/how-to-find-and-replace-a-part-of-a-value-in-json-file

with open('book.json') as f:
    data = json.load(f)

for item in data['Books']:
    # it would not accept teh fada in my name
    item['Author'] = item['Author'].replace('Andrew Beatty', 'Faolan Hamilton')
# w allows for the addition above to be written into a new file
with open('new_book.json', 'w') as f:
    json.dump(data, f)
```
