```python
# Source: https://stackoverflow.com/questions/15138614/how-can-i-read-the-contents-of-an-url-with-python#:~:text=Copy-,import%20requests,-link%20%3D%20%22http
# The API url for the deck of cards
card_deck = "https://deckofcardsapi.com"

# Shuffling the deck
deck_shuffled = "https://deckofcardsapi.com/api/deck/new/shuffle/?deck_count=1"

# "Dealing out" five cards
deal_out = "https://deckofcardsapi.com/api/deck/gnhx64eb4mx9/draw/?count=5"

five_cards = requests.get(deal_out).json()
```

```python
# Outlining the values and suits
for five_card in five_cards["cards"]:
    print(f"{five_card["value"]} of {five_card["suit"]}")
```

```python
cards = five_cards["cards"]

values  = [card["value"] for card in cards]
suits   = [card["suit"] for card in cards]
```

```python
value_counts = Counter(values)

# Pair
has_pair = any(count == 2 for count in value_counts.values())

# Three of a kind
has_three = any(count == 3 for count in value_counts.values())

# A Flush (A hand of five cards with the same suit - source https://www.mypokercoaching.com/poker-hands-rankings/flush/)
is_flush = len(set(suits)) == 1

# A straight - consecutive cards in order e.g. 2,3,4,5,6
rank_map = {
    "ACE": 14, "KING": 13, "QUEEN": 12, "JACK": 11,
    "10": 10, "9": 9, "8": 8, "7": 7, "6": 6,
    "5": 5, "4": 4, "3": 3, "2": 2
}

ranks = sorted(rank_map[v] for v in values)

ace_low = [14 if r == 14 else r for r in ranks]  # treat Ace as 14
ace_low = sorted([1 if r == 14 else r for r in ranks])
is_straight = (
    all(ranks[i] + 1 == ranks[i+1] for i in range(4)) or
    all(ace_low[i] + 1 == ace_low[i+1] for i in range(4))
)
```

```python
print("Your hand:")
for c in cards:
    print(f"{c['value']} of {c['suit']}")

print("\nHand results:")

if has_three:
    print("Congratulations - You have a three-of-a-kind!")
elif has_pair:
    print("Congratulations - You have a pair!")

if is_straight:
    print("Congratulations - You have a straight!")

if is_flush:
    print("Congratulations - You have a flush!")

if not (has_pair or has_three or is_straight or is_flush):
    print("No special hand this time, better luck next time :().")
```
